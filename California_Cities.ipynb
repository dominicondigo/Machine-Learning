```python
#confirming the ten lowest values of latitude values
from heapq import nsmallest
lst = data['latd']
lowest = nsmallest(10, lst)
lowest
```

```python
#We can't use heapq to find the min, max because it's a list 
#here will mean, max, min, std, medium of the ten lowest values of latitude values	
#data.nsmallest(10, 'latd').describe()
data['latd'].nsmallest(10).describe()
```

```python
#4.	Display the ten cities with the least value of elevation in Meters
data.nsmallest(10,'elevation_m')
```

```python
#5. top ten cities with the highest population totals
data.nlargest(10,'population_total')
```
